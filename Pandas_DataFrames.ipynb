##Operations

**Operations in series** work at a column level, broadcasting down the rows(Which can be counter intutive)

```python
crisis = pd.Series([-1000000,-0.3], index=['GDP','HDI'])
```

```python
crisis
```

```python
df[['GDP','HDI']]
```

```python
df[['GDP','HDI']] + crisis
```

#Modifyng DataFrames

###Adding a new column

```python
langs = pd.Series(['French','German','Italian'],
                  index=['France','Germany','Italy'],
                  name='Language')
```

```python
langs
```

```python
df['Language'] = langs
```

```python
df
```

```python
df['Language']='English'
```

```python
df
```

```python
df['Language'] = ['English','French','German','Italian','Japanese','English','English']
```

```python
df
```

###Renaming columns

```python
df.rename(
    columns={
        'HDI':'Human Development Index'
    },
    index={
        'United Kingdom':'UK',
        'United States':'US'
    }
)
```
